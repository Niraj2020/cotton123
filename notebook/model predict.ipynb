# VGG16 Model Prediction

```python
from tensorflow.keras.preprocessing.image import load_img
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from tensorflow.keras.preprocessing import image
```

```python
import cv2
import tensorflow as tf
from tensorflow.keras.models import load_model
```

```python
model = load_model('model_vgg16.h5')
```

```python
test_image=image.load_img('Datasets/test/fresh cotton leaf/d (133)_iaip.jpg', target_size=(224, 224))
```

```python
imgplot=plt.imshow(test_image)
```

```python
test_image=image.img_to_array(test_image)
```

```python
test_image=test_image/255
```

```python
test_image=np.expand_dims(test_image, axis=0)
preds=model.predict(test_image)
```

```python
preds
```

```python
preds=np.argmax(preds, axis=1)
```

```python
if preds==0:
    print("The leaf is diseased cotton leaf")
elif preds==1:
    print("The leaf is diseased cotton leaf")
elif preds==2:
    print("The leaf is fresh cotton leaf")
else:
    print("The leaf is fresh cotton plant")
    
    
```
